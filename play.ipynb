```python
import numpy as np
```

```python
probs = np.random.rand(100, 10)
labels = np.random.randint(0, 10, 100)
```

```python
confidences = probs.max(axis=1)
predictions = probs.argmax(axis=1)
correct = (predictions == labels).astype(float)

bin_edges = np.linspace(0.0, 1.0, 15 + 1)
bin_confidences = np.zeros(15)
bin_accuracies = np.zeros(15)
bin_counts = np.zeros(15, dtype=int)

for i in range(15):
    lo, hi = bin_edges[i], bin_edges[i + 1]
    mask = (confidences > lo) & (confidences <= hi) if i > 0 else (confidences >= lo) & (confidences <= hi)
    count = mask.sum()
    bin_counts[i] = count
    if count > 0:
        bin_confidences[i] = confidences[mask].mean()
        bin_accuracies[i] = correct[mask].mean()

```
